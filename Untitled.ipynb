```python
import os
filePath = 'Mello\\'
filename_list=os.listdir(filePath)
print(filename_list)
```

```python
dataframe_lists=[]
```

```python
for file in filename_list:
    if file[-4:]=="xlsx":
        df = pd.read_excel(filePath+file)
        dataframe_lists.append(df)
```

```python
地理位置=[]
电信公司=[]
```

```python
def ge_address_company(x):
    slash_index=int(x.find('-'))
#     print(slash_index)
    地理位置.append(x[0:slash_index])
#     print(x[0:slash_index])
    电信公司.append(x[slash_index+1:])
#     print(x[slash_index:])
#     print(x[slash_index])
```

```python
result['地区'].map(lambda x :ge_address_company(x));
```

```python
result['地理位置']=地理位置
```

```python
result['电信公司']=电信公司
```

```python
date_range=["25","26","27","28","29","30","31"]
```

```python
for date in date_range:
    print("开始换日期")
    calendar=wd.find_element_by_class_name("btn.btn-sm.btn-secondary.mr-3.vote-item-chart-date")
    calendar.click()
    sleep(1)
    content=wd.find_element_by_class_name('layui-laydate-content')# 找到日历的那个框
    date_eles=content.find_elements_by_tag_name('td')# 从日历的那个框里面找对应的日期
    data_pre=[]#准备空的，有可能会有两个相同日期
    for i in date_eles:
        if i.text==date:
            data_pre.append(i)

    if int(date)>15:#如果大于15，有可能前面会有重复，一般选倒数第一个
        date_right=data_pre[-1]
    elif int(date)<=15:#如果小于15， 那肯定是第一个
        date_right=data_pre[0]
    
    date_right.click()#选中之后点击换日期
    print(f"换日期中 换到  {date}")
    sleep(5)

    
    while True:
        print("看看有没有")
        footer=wd.find_elements_by_class_name('btn.btn-sm.btn-pill.btn-secondary')
        break_flag=0
        for i in footer:
            if i.text=="下一页":
                print("找到下一页啦")
                excel_output_button=wd.find_element_by_class_name('btn.btn-sm.btn-info')#找到导出本页为excel的按钮
                excel_output_button.click()
                print("下载中")
                i.click()
                print("下一页")
                sleep(4)
                break_flag+=1
        if break_flag==0:
            print("该换日期了")
            break
                
```
